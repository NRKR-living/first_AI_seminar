## Implimenting K-NN Class

K-NNを実装する。K-meanのときはkが「クラスタの数」を表していたが、k-NNでは「近傍のデータ点何個をラベル付けの参考にするか」であることに注意。

また、後でsklearnの便利関数（GridSearchCV）を使いたいので、sklearnの抽象基底クラスに最低限合わせたメソッドをもたせている。  
具体的には以下の機能をもたせたクラスが「sklearnの抽象基底クラスに最低限合わせた実装」である。
- \__init\__
    - データは受け取らず、学習器のハイパーパラメータのみを渡し、何もreturnしない。（モデルが訓練の最中に変更していく値をパラメータといい、学習前に人が与える設定値をハイパーパラメータという）
- fit
    - 訓練メソッド。X_trainとy_trainを受け取って訓練を行い、訓練した結果パラメータが変更された自分自身を返す（return self）。
- predict
    - 予測メソッド。X_testのみ受け取って予測したラベルを返す。
- score
    - X_testとy_testを受け取って、predictしてから正答率などの評価指標で評価した値を返す。
- set_params
    - \__init__の引数で定義できるハイパーパラメータをここからもセットできるようにしている。return selfする。
- get_params
    - そのモデルにセットされたハイパーパラメータの辞書を返す。
    
これらの機能を持ったクラスであれば、sklearnの便利な関数を利用する事ができる（かもしれない）ので、自分で実装するときにも是非合わせてほしい。

```python
class BaseClassifier:
    def fit(self, X_train, y_train):
        raise NotImplementedError
    
    def predict(self, X_test):
        raise NotImplementedError
    
    def score(self, X, y):
        pred_labels = self.predict(X)
        return self.acc(pred_labels, y)
    
    def set_params(self, **parameters):
        for parameter, value in parameters.items():
            setattr(self, parameter, value)
        return self

    def get_params(self, deep=True):
        raise NotImplementedError

class KNN(BaseClassifier):
    def __init__(self,k=1):
        self.k = k
        
    def fit(self, X_train:np.ndarray, y_train:np.ndarray):
        self.X_train = X_train
        self.y_train = y_train
        return self
    
    def predict(self, X_test:np.ndarray, distance_fn=lambda a,b: np.sum((a-b)**2, axis=-1))->np.ndarray:
        """ distance_fn: default is sum of squares 
        （ここではユークリッド距離の代わりに2乗和を使ったが、全ての距離からルートを取っただけなので大小関係は同じなので問題はない）
        """
        predicted_labels = []
        for x in X_test:
            distances = distance_fn(x, self.X_train)
            sorted_index = np.argsort(distances)
            neighbors = self.y_train[sorted_index][:self.k]
            labels_freq = np.bincount(neighbors)
            # [2,1,2]
            predicted_labels.append(np.argmax(labels_freq))
        return np.array(predicted_labels)
    
    def get_params(self, deep=True):
        return {"k": self.k}
    
    @staticmethod 
    def acc(y_pred:np.array, y_true:np.array)->np.float64:
        """acc (accuracy, 正答率)
        """
        return (y_pred == y_true).sum() / y_true.shape[0]
```

accメソッドの上についた@staticmethodという記述に注目。  
これのように、メソッドの上につける@hogehogeのようなものは__デコレータ__という。デコレータはPython特有の機能。簡単に言うと、デコレータは、メソッドに値が入る前とメソッド処理後の２つの時点で何かしらの処理を挿入できる機能。
 
例えばこの@staticmethodはメソッドをクラスの外からでも簡単に使えるようにする。これがついたメソッドはselfが不要（逆に言うとselfが参照できない）。 わかりやすく「@staticmethodはメソッドをただの関数にする機能」だと思っても（今は）良い。

### Grid Search about K

k=2~ 訓練セットのデータ数 の範囲でKnnの性能を評価してみよう。
GridSearchCVクラスはscoreメソッドの返り値が最も良くなるハイパーパラメータの組み合わせを探索してくれる。

- 第1引数：最適化したいモデル
- 第2引数：探索したいパラメータ空間を持った辞書
- cv: Cross Validationするときに何分割して値を計算するかを指定する値

モデルの最適なハイパーパラメータを探すときには、テストセットを利用することができない。何故ならテストセットに対してKの値を最適化させたら「テストセットの情報を使って学習した」こととほぼ同義になってしまい、モデルの汎化性能に疑問符がつくから。

そのため訓練セットの中の一部をテストデータの代わりとして使うことにする。これをバリデーションセット（バリデーションデータ、評価データとも）という。つまり自由に使えるデータ（今まで訓練セットと呼んでいたもの）を更に、訓練セットとバリデーションセットに分ける。cv=5ならば「訓練セットを5分割して、一つをバリデーションセットとして使う」という意味。この分割したサブセットをfoldとよぶ。なのでcv=5ならば5-fold cross validationと呼ばれる。

また、五個に分割したのでそれぞれがバリデーションセットの場合の評価値をまとめて平均値をとって、これが俗にCVスコアとか呼ばれている値になる。こうしてバリデーションセットを何個か作って評価値を平均することで、一部を最初にホールドアウトして評価しただけの場合よりも汎化性能が高いモデルが得られる。

これについては以下が分かりやすい。
https://aizine.ai/cross-validation0910/

仮定の通り、ACCはあまり良い値になっていない。それでは次に、正規化（全ての特徴を0~1の範囲に収める処理）を行ってからK-NNにかけてみる。仮定が正しければ先程よりは高いACCが出るはずだ。

```python
def normalize(x, axis=None):
    min = x.min(axis=axis, keepdims=True)
    max = x.max(axis=axis, keepdims=True)
    result = (x-min)/(max-min)
    return result
```

```python
nomalized_X_train = normalize(X_train, axis=0)
params = {"k":range(2,20)} 
gs = GridSearchCV(KNN(), params, cv=5)
gs.fit(nomalized_X_train, y_train)
```

```python
result_df = pd.DataFrame.from_dict(gs.cv_results_)
result_df
```

```python
result_df = result_df.T 
num_experiments = result_df.shape[-1]
accuracies = [result_df.loc["mean_test_score"][i] for i in range(num_experiments)]
plt.plot(params["k"], accuracies, )
plt.title("Validation Accuracy")
plt.xlabel("K")
plt.ylabel("ACC")
```

```python
print(f"Best Parameters: {gs.best_params_}\nBest Validation ACC Score: {gs.best_score_}")
```

```python
# best model

best_model = gs.best_estimator_
normalized_X_test = normalize(X_test, axis=0)
test_acc = best_model.score(normalized_X_test, y_test)

print(f"Best models's K: {best_model.k}")
print(f"Test set's accuracy: {test_acc}")
```

期待したとおり、ACCは最初より高い値になった。注意しなければならないのは、ここで行ったnormalizationは外れ値に大きな影響を受けるということ。大多数のデータと離れた位置にデータ一つでもあった場合は、その外れ値を0~1の範囲に収めるために大多数のデータが狭い範囲にマップされる。凡そいい影響は与えないだろうから、そのような場合は別の前処理を検討する必要がある。

# appendix 2

Pure Python（Numpyを使わない） K-NNの実装


```python
def euclid_dist(a, B):
    l2 = []
    for b in B:
        tmp = []
        for i in range(len(b)):
            tmp.append((a[i]-b[i])**2)
        l2.append(sum(tmp)**(1/2))
    return l2
    
def bincount(items):
    #print(items)
    n_kind = max(items)+1 # start index is 0
    counters = [0 for i in range(n_kind)]
    for item in items:
        counters[item] += 1
    return counters

def argsort(items, reverse=False):
    _index = list(range(len(items)))
    sorted_index = sorted(_index, key=lambda i: items[i])
    if reverse:
        return sorted_index
    else:
        return list(reversed(sorted_index))
    
def argmax(items):
    sorted_index = argsort(items, reverse=True)
    return sorted_index[-1]

class KNN_python(BaseClassifier):
    def __init__(self,k=1):
        self.k = k
        
    def fit(self, X_train:np.ndarray, y_train:np.ndarray):
        self.X_train = X_train
        self.y_train = y_train
        return self
    
    def predict(self, X_test:np.ndarray, distance_fn=euclid_dist)->np.ndarray:
        """ distance_fn: default is sum of squares 
        （ここではユークリッド距離の代わりに2乗和を使ったが、全ての距離からルートを取っただけなので大小関係は同じなので問題はない）
        """
        predicted_labels = []
        for x in X_test:
            distances = distance_fn(x, self.X_train)
            sorted_index = argsort(distances)
            neighbors = [self.y_train[i] for i in sorted_index[:self.k]]
            
            labels_freq = bincount(neighbors)
            predicted_labels.append(argmax(labels_freq))
        return np.array(predicted_labels)
    
    def get_params(self, deep=True):
        return {"k": self.k}
    
    @staticmethod 
    def acc(y_pred:np.array, y_true:np.array):
        """acc (accuracy, 正答率)
        """
        n_items = len(y_pred)
        counter = 0
        for i in range(n_items):
            if y_pred[i] == y_true[i]:
                counter += 1
        return counter/ n_items
```

```python
model = KNN_python(k=8)
model.fit(X_train,y_train)
pred_labels = model.predict(X_test)
model.acc(pred_labels, y_test)
```
